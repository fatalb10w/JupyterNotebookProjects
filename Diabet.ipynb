```python
#use the .NAME file to view and set the features of the dataset
feature_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'flag']
```

```python
#using the feature names set earlier and fix it as the column headers of the dataset
df_diabets_data_2 = pd.read_csv("C:\Project1\[SW.BAND] pima-indians-diabetes.data", header = None, names = feature_names)
```

```python
#verifying if the dataset is updated with the new headers
df_diabets_data_2.head()
```

```python
#the number of observation and features of the dataset 
df_diabets_data_2.shape
```

```python
#selecting features from the dataset to create the model
feature_selected = ['pregnant', 'insulin', 'bmi', 'age']
```

```python
#creating the feature object
X_features = df_diabets_data_2[feature_selected]
```

```python
#creating the response object
Y_target = df_diabets_data_2['flag']
```

```python
#view the shape of the feature object
X_features.shape
```

```python
#view the shape of the target object
Y_target.shape
```

```python
#split the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_features, Y_target, random_state=27)
```

```python
#creating a logistic regression model using the training set
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
#making predictions using the testing set
y_pred = logreg.predict(x_test)
```

```python
#the accuracy of model
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
