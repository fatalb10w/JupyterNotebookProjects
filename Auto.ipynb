```python
#import libraries
import pandas as pd
import seaborn as sns
```

```python
df_auto = pd.read_csv("C:\Project1\\auto_data.csv")
```

```python
df_auto.head()
```

```python
#write a user defined function for origin
# 1 - USA, 2 - Europe, 3 - Asia
def origin_conversion(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'
```

```python
#use apply function
df_auto['origin'] = df_auto['origin'].apply(origin_conversion)
```

```python
df_auto.head(30)
```

```python
#create pairplot
sns.pairplot(df_auto[['mpg', 'weight', 'origin']], hue = 'origin', height = 2.5)
```
