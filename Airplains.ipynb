```python
#create new dataset
for_analysis_df_faa = df_faa[[ 'ACFT_MAKE_NAME', 'LOC_STATE_NAME', 'ACFT_MODEL_NAME', 'RMK_TEXT', 'FLT_PHASE', 'EVENT_TYPE_DESC', 'FATAL_FLAG']]
```

```python
for_analysis_df_faa['FATAL_FLAG'].fillna(value='No', inplace = True)
```

```python
final_table = for_analysis_df_faa.dropna(subset = ['ACFT_MAKE_NAME'])
```

```python
final_table.shape
```

```python
#group by aircraft name
aircraft = final_table.groupby(['ACFT_MAKE_NAME'])
```

```python
#view aircraft count
aircraft.size()
```

```python
#group by fatal flag
aircraft_fatal_flag = final_table.groupby(['FATAL_FLAG'])
```

```python
#view fatal flag count
aircraft_fatal_flag.size()
```

```python
#create new table with fatal accidents
fatal_accidents = aircraft_fatal_flag.get_group(('Yes',))
```
